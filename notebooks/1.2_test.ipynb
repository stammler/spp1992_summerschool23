# Testing

The purpose of this notebook is to test the installation. Please run all cells without any further explanation. A detailed explanation will follow during the tutorial.  
The execution of all cells will take a couple of minutes.

```python
from dustpy import plot
from dustpy import Simulation
from dustpy import constants as c
from dustpylib.radtrans import radmc3d
import numpy as np
```

## Testing DustPy

```python
s = Simulation()
s.initialize()
s.t.snapshots = np.geomspace(1.e2, 1.e3, 2) * c.year
s.writer.datadir = "test"
s.run()
```

```python
plot.panel(s)
```

## Testing DustPyLib

```python
rt = radmc3d.Model(s)
rt.datadir = "test_radmc3d"
rt.write_files()
```
